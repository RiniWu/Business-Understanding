###**Outliers**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Melakukan Imputasi pada kolom 'Age',
 'Latitude', 'Longitude'.

```python
columns_to_impute = ["Age", "latitude", "longitude"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

Pengecekan Ulang

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

#**VISUALISASI DATA**

##**1. Distribusi Usia Pelanggan**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Age"], bins=20, kde=True, color="blue")
plt.title("Distribusi Usia Pelanggan")
plt.xlabel("Usia")
plt.ylabel("Jumlah Pelanggan")
plt.show()
```

Visualisasi di atas menunjukkan distribusi usia pelanggan yang menggunakan layanan pemesanan makanan online. Dari grafik ini, kita dapat melihat apakah ada kelompok usia tertentu yang mendominasi penggunaan layanan.

Insight dan Tindakan yang Bisa Diambil:

1. Jika mayoritas pengguna berada di kelompok usia tertentu (misalnya 20-30 tahun), strategi pemasaran bisa difokuskan pada preferensi dan kebiasaan konsumsi mereka.
2. Jika distribusi cukup merata, bisa dilakukan segmentasi yang lebih spesifik berdasarkan faktor lain seperti pendapatan atau preferensi makanan.

##**2. Preferensi Makanan berdasarkan Gender**

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Preferred Food Type", hue="Gender", palette="Set2")
plt.title("Preferensi Makanan Berdasarkan Gender")
plt.xlabel("Jenis Makanan")
plt.ylabel("Jumlah Pelanggan")
plt.xticks(rotation=45)
plt.legend(title="Gender")
plt.show()
```

Grafik di atas menunjukkan preferensi makanan berdasarkan gender. Dari sini, kita dapat melihat pola konsumsi yang berbeda antara pelanggan pria dan wanita.

Insight dan Tindakan yang Bisa Diambil:

1. Jika terdapat perbedaan signifikan dalam preferensi makanan berdasarkan gender, kampanye pemasaran dapat disesuaikan. Misalnya, jika wanita lebih banyak memilih makanan vegan, promosi khusus bisa ditargetkan ke segmen ini.
2. Jika ada makanan yang lebih disukai oleh kedua gender, maka menu tersebut bisa dijadikan rekomendasi utama untuk meningkatkan penjualan.

##**3. Pembelian berdasarkan Perndapatan Bulanan**

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="Monthly Income", hue="Output", palette="coolwarm")
plt.title("Keputusan Pembelian Berdasarkan Pendapatan Bulanan")
plt.xlabel("Pendapatan Bulanan")
plt.ylabel("Jumlah Pelanggan")
plt.xticks(rotation=45)
plt.legend(title="Melakukan Pembelian")
plt.show()
```

Grafik di atas menunjukkan hubungan antara pendapatan bulanan pelanggan dan keputusan mereka untuk melakukan pembelian.

Insight dan Tindakan yang Bisa Diambil:

1. Jika pelanggan dengan pendapatan tertentu lebih cenderung melakukan pembelian, strategi harga dan promosi dapat difokuskan pada segmen ini.
2. Jika pelanggan dengan pendapatan rendah jarang melakukan pembelian, perusahaan bisa menawarkan diskon atau paket hemat untuk menarik lebih banyak pesanan dari kelompok ini.
3. Jika pelanggan tanpa pendapatan (misalnya, mahasiswa) cukup aktif membeli, promosi khusus untuk segmen ini bisa lebih dioptimalkan.
